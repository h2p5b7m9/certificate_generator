## Opción certificados con nombre, fecha y token como en la app

```python
from PIL import Image, ImageDraw, ImageFont
import datetime


def generate_certificate(certificated_to,user):
  font = ImageFont.truetype('OpenSans-Bold.ttf',60)
  token= ImageFont.truetype('OpenSans-Bold.ttf',20)
  date= ImageFont.truetype('OpenSans-Bold.ttf',25)

  img = Image.open('media/certificate_templates/'+certificated_to.slug+'.jpg')
  img2 = img
  draw = ImageDraw.Draw(img)
  draw2 = ImageDraw.Draw(img2)
  draw.text(xy=(200,390),text='{}'.format(user),fill=(105, 96, 96),font=font)
  draw.text(xy=(1820,1360),text='{}'.format(datetime.datetime.strftime("%d/%m/%y")),fill=(105, 96, 96),font=date)
  draw.text(xy=(1600,50),text='{}'.format(token),fill=(175,160,160),font=token)

  draw2.text(xy=(200,390),text='{}'.format(user),fill=(105, 96, 96),font=font)
  draw2.text(xy=(1820,1360),text='{}'.format(datetime.datetime.strftime("%d/%m/%y")),fill=(105, 96, 96),font=date)
  draw2.text(xy=(1600,50),text='{}'.format(token),fill=(175,160,160),font=token)

  img.save('media/certificates/{}.pdf'.format(token))
  img2.save('media/certificates/{}.jpg'.format(token))
```

# Automatizar envío por email si tenemos email

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.application import MIMEApplication

def send_certificate(email, username):
    sender_email = "tucorreo@example.com"
    receiver_email = email
    password = "tucontraseña"

    # Configura el servidor SMTP y el puerto
    # Por ejemplo, usando Gmail como servidor SMTP
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login(sender_email, password)

    # Crea el mensaje de correo electrónico
    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = receiver_email
    msg['Subject'] = "Tu Certificado"

    body = "Hola {}, aquí está tu certificado.".format(username)
    msg.attach(MIMEText(body, 'plain'))

    # Adjunta el archivo PDF
    filename = f'media/certificates/certificado_21_days_{username}.pdf'
    attachment = open(filename, "rb")
    part = MIMEApplication(attachment.read(), Name=filename)
    part['Content-Disposition'] = f'attachment; filename="{filename}"'
    msg.attach(part)

    # Envía el correo
    server.send_message(msg)
    server.quit()
    attachment.close()

# Usar la función para enviar un certificado
send_certificate("destinatario@example.com", "John Doe")
```

```python
# autenticar en google drive
from google.colab import auth
auth.authenticate_user()

from googleapiclient.discovery import build
drive_service = build('drive', 'v3')
sheets_service = build('sheets', 'v4')
```

```python
# subir archivos
from googleapiclient.http import MediaFileUpload

def upload_to_drive(filename, filepath):
    # filepath-> ruta del archivo
    # nombre de archivo-> name:filename
    # tipo de archivo -> mimeType:'application/pdf'
    file_metadata = {
        'name': filename,
        'mimeType': 'application/pdf'
    }
    # crea objeto media para cargar archivos
    media = MediaFileUpload(filepath, mimetype='application/pdf')
    # Esta línea usa el servicio de Google Drive (drive_service) para crear un nuevo archivo en Drive
    file = drive_service.files().create(body=file_metadata, media_body=media, fields='id').execute()
    # esta línea extrae el ID del archivo de la respuesta de la API.
    file_id = file.get('id')

    # Generar un link compartible
    # Permisos del archivo en modo lectura con el link
    drive_service.permissions().create(fileId=file_id, body={'type': 'anyone', 'role': 'reader'}).execute()
    # forma el URL que permite acceder al archivo usando el file_id
    link = f"https://drive.google.com/file/d/{file_id}/view"
    return link
```
